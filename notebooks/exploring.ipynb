```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
qp_filepath = "../data/qp.csv"
df = pd.read_csv(qp_filepath)
df
```

```python
df.groupby("nom_commune").sum().sort_values("pop_mun", ascending=False)
```

```python
fqp_filepath = "../data/foncier_qp.csv"
dff = pd.read_csv(fqp_filepath)
dff
```

```python
tdf = dff.groupby("nom_commune")["valeur_fonciere", "surface_reelle_bati", "surface_terrain"].mean()
tdf["pm2"] = tdf.valeur_fonciere / tdf.surface_reelle_bati
tdf
```

```python
tdf.pm2.sort_values().plot.bar(figsize=(16,5))
```
